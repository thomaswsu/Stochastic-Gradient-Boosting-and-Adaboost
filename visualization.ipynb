```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
#model converence.
num_itr = 100
model_loss = [1/i for i in range(1, num_itr+1)]

plt.title("Model Converence")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid()
plt.plot([i for i in range(num_itr)], model_loss, label="Gradient Boosting")
plt.plot([i for i in range(num_itr)], model_loss, label="Stochastic Gradient Boosting")

plt.plot([i for i in range(num_itr)], model_loss, label="Gradient Descent")
plt.plot([i for i in range(num_itr)], model_loss, label="Stochastic Gradient Descent")
plt.legend()

plt.show()
```

```python
#accuracy converence.
num_itr = 100
training_acc = [5**-1 * np.log(i) for i in range(1, num_itr+1)]
testing_acc = [10**-1 * np.log(i) for i in range(1, num_itr+1)]

plt.figure(1,figsize=(20,5))

plt.subplot(131)
plt.title("Regular Model Accuracy")
plt.xlabel("Iteration")
plt.ylabel("Accuracy")
plt.grid()
plt.plot([i for i in range(num_itr)], training_acc, label="GD - Training")
plt.plot([i for i in range(num_itr)], testing_acc, label="GD - Testing")

plt.plot([i for i in range(num_itr)], training_acc, label="GB - Training")
plt.plot([i for i in range(num_itr)], testing_acc, label="GB - Testing")
plt.legend()

plt.subplot(132)
plt.title("Stochastic Model Accuracy")
plt.xlabel("Iteration")
plt.ylabel("Accuracy")
plt.grid()
plt.plot([i for i in range(num_itr)], training_acc, label="SGD - Training")
plt.plot([i for i in range(num_itr)], testing_acc, label="SGD - Testing")

plt.plot([i for i in range(num_itr)], training_acc, label="SGB - Training")
plt.plot([i for i in range(num_itr)], testing_acc, label="SGB - Testing")
plt.legend()

plt.show()
```

```python
import csv

def data_formating(losses, accuracies):
    f_data = []
    
    for i in range(len(losses[0])):
        temp = []
        for j in range(len(losses)):
            temp.append(losses[j][i])
            temp.append(accuracies[j][i])
        f_data.append(temp)
    
    return f_data


def csv_formater(file_name, lower, higher, losses, accuracies):
    file_name = file_name + ".csv"
    with open(file_name, 'w', newline='') as csvfile:
        fieldnames = []
        steps = [i/10 for i in range(lower, higher)]
        for i in range(len(steps)):
            fieldnames.append("Loss: {}".format(steps[i]))
            fieldnames.append("Training Accuracy: {}".format(steps[i]))

        writer = csv.writer(csvfile) 

        writer.writerow(fieldnames) 

        data = data_formating(losses, accuracies)
        writer.writerows(data)
    
```

```python
num_itr = 100
model_loss_1 = training_acc_1 = [i for i in range(1, num_itr+1)]
model_loss_2 = training_acc_2 = [i+1 for i in range(1, num_itr+1)]
model_loss_3 = training_acc_3 = [i+2 for i in range(1, num_itr+1)]

losses = [model_loss_1, model_loss_2, model_loss_3]
accuracies = [training_acc_1, training_acc_2, training_acc_3]

csv_formater("test", 1, 4, losses, accuracies)
```
